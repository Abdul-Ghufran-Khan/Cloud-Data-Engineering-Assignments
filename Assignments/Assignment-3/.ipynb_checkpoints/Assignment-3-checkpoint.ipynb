```python
# Problem 1: Count Vowels
input_string = input("Enter a string: ")
vowels = "aeiouAEIOU"
vowel_count = 0
for char in input_string:
    if char in vowels:
        vowel_count += 1
print("Number of vowels:", vowel_count)
```

```python
# Problem 2: Character and Whitespace Count
input_string = input("Enter a string: ")
upper_count = 0
lower_count = 0
digit_count = 0
whitespace_count = 0

for char in input_string:
    if 'a' <= char <= 'z':
        lower_count += 1
    elif 'A' <= char <= 'Z':
        upper_count += 1
    elif '0' <= char <= '9':
        digit_count += 1
    elif char == ' ':
        whitespace_count += 1
        
print("Number of uppercase letters:", upper_count)
print("Number of lowercase letters:", lower_count)
print("Number of digits:", digit_count)
print("Number of whitespace characters:", whitespace_count)
```

```python
# Problem 3: Swap First and Last Characters
s = input("Enter a string: ")
if len(s) < 2:
    new_s = s
else:
    new_s = s[-1] + s[1:-1] + s[0]
print("New string:", new_s)
```

```python
# Problem 4: Reverse First and Last
s = input("Enter a string: ")
if len(s) < 2:
    print("New string:", s)
else:
    s_list = list(s)
    first_char = s_list[0]
    last_char = s_list[-1]
    s_list[0] = last_char
    s_list[-1] = first_char
    new_s = "".join(s_list)
    print("New string:", new_s)
```

```python
# Problem 5: Shift String by One Position
s = input("Enter a string: ")
if not s:
    new_s = ""
else:
    new_s = s[1:] + s[0]
print("New string:", new_s)
```

```python
# Problem 6: Print Initials
full_name = input("Enter your full name: ")
parts = full_name.strip().split()
initials = ""
for part in parts:
    initials += part[0].upper() + ". "
print("Initials:", initials.strip())
```

```python
# Problem 7: Palindrome Checker
s = input("Enter a string: ")
clean_s = s.replace(" ", "").lower()
if clean_s == clean_s[::-1]:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
# Problem 8: Display Output
word = "SHIFT"
length = len(word)
for i in range(length+1):
    shifted_word = word[i:] + word[:i]
    print(shifted_word.upper())
```

```python
#### Problem 9: Password Setup
import re

password = input("Enter a password: ")
is_valid = True

if len(password) < 8:
    print("Password must be at least 8 characters long.")
    is_valid = False

if not re.search("[A-Z]", password):
    print("Password must contain at least one uppercase letter.")
    is_valid = False

if not re.search("[a-z]", password):
    print("Password must contain at least one lowercase letter.")
    is_valid = False
    
if not re.search("[0-9]", password):
    print("Password must contain at least one digit.")
    is_valid = False
    
if not re.search("[!@#$%^&*()_+={}\[\]|\\:;\"'<,>.?/`~-]", password):
    print("Password must contain at least one special character.")
    is_valid = False

if is_valid:
    print("Password is valid.")
```
