```python
try:
    user_input = input("Enter a list of numbers, separated by spaces: ").split()
    my_list = [int(x) for x in user_input]
    if not my_list:
        print("The list is empty.")
    else:
        largest = my_list[0]
        for number in my_list:
            if number > largest:
                largest = number
        print("The largest number is:", largest)
except ValueError:
    print("Invalid input. Please enter numbers only.")
```

```python
my_list = input("Enter a list of elements, separated by spaces: ").split()
rotated_list = my_list[:] # Create a copy
for i in range(0, len(rotated_list) - 1, 2):
    rotated_list[i], rotated_list[i+1] = rotated_list[i+1], rotated_list[i]
print("Rotated list:", rotated_list)
```

```python
my_string = input("Enter a string: ")
word_to_delete = input("Enter the word to delete: ")
new_string = my_string.replace(word_to_delete, "").strip()
new_string = " ".join(new_string.split()) # To handle extra spaces
print("New string:", new_string)
```

```python
from datetime import datetime

user_date = input("Enter a date (dd/mm/yyyy): ")
try:
    date_object = datetime.strptime(user_date, "%d/%m/%Y")
    formatted_date = date_object.strftime("%B %d, %Y")
    print("Formatted date:", formatted_date)
except ValueError:
    print("Invalid date format. Please use dd/mm/yyyy.")
```

```python
matrix = [
    [2, 11, 7, 12],
    [5, 2, 9, 15],
    [8, 3, 10, 42]
]
for i, row in enumerate(matrix):
    row_sum = sum(row)
    print(f"Sum of row {i + 1} = {row_sum}")
```

```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
result_matrix = []

# Check if matrices can be added (same dimensions)
if len(A) == len(B) and len(A[0]) == len(B[0]):
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        result_matrix.append(row)
    
    print("Sum of the two matrices:")
    for row in result_matrix:
        print(row)
else:
    print("Matrices must have the same dimensions for addition.")
```

```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

# Check if matrices can be multiplied
if len(A[0]) == len(B):
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])

    result_matrix = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result_matrix[i][j] += A[i][k] * B[k][j]
    
    print("Product of the two matrices:")
    for row in result_matrix:
        print(row)
else:
    print("Matrix multiplication is not possible. Columns of the first matrix must equal rows of the second.")
```
